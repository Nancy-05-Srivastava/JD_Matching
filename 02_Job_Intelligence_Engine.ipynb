```python
import json
import re
import numpy as np
import pandas as pd

from collections import Counter
```

```python
feature_df = pd.read_csv("candidate_features.csv")

feature_df.head()
```

```python
JOB_DESCRIPTION = """
Paste the complete job description here.
"""
```

```python
jd = JOB_DESCRIPTION.lower()

jd = re.sub(r"\s+"," ",jd)
```

```python
TECH_STACK = {

"languages":[
"python",
"java",
"scala",
"go",
"c++"
],

"ml":[
"machine learning",
"deep learning",
"pytorch",
"tensorflow",
"xgboost",
"lightgbm"
],

"llm":[
"llm",
"transformers",
"bert",
"llama",
"rag",
"embedding",
"retrieval",
"vector",
"reranking",
"sentence transformers"
],

"infra":[
"docker",
"kubernetes",
"aws",
"azure",
"gcp",
"spark",
"kafka",
"airflow"
]

}
```

```python
jd_skills=[]

for category,skills in TECH_STACK.items():

    for skill in skills:

        if skill in jd:

            jd_skills.append(skill)

jd_skills
```

```python
experience_pattern = r"(\d+)\+?\s*years"

matches = re.findall(
    experience_pattern,
    jd
)

matches
```

```python
mandatory = [

"must",

"required",

"strong",

"expert",

"mandatory"

]

mandatory_sentences=[]

for sentence in JOB_DESCRIPTION.split("."):

    for word in mandatory:

        if word in sentence.lower():

            mandatory_sentences.append(sentence.strip())

mandatory_sentences
```

```python
preferred = [

"preferred",

"nice to have",

"good to have",

"plus"

]

preferred_sentences=[]

for sentence in JOB_DESCRIPTION.split("."):

    for word in preferred:

        if word in sentence.lower():

            preferred_sentences.append(sentence.strip())

preferred_sentences
```

```python
negative=[]

phrases=[

"not",

"avoid",

"without",

"instead of"

]

for sentence in JOB_DESCRIPTION.split("."):

    for word in phrases:

        if word in sentence.lower():

            negative.append(sentence.strip())

negative
```

```python
production_keywords=[

"production",

"distributed",

"scalable",

"real time",

"pipeline",

"deployment",

"latency",

"serving",

"retrieval",

"ranking",

"evaluation"

]

production=[]

for word in production_keywords:

    if word in jd:

        production.append(word)

production
```

```python
startup_keywords=[

"ownership",

"startup",

"fast",

"iterate",

"ship",

"product",

"customer",

"impact"

]

startup=[]

for word in startup_keywords:

    if word in jd:

        startup.append(word)

startup
```

```python
weights={

"semantic_similarity":35,

"production_experience":25,

"behavioral_signals":15,

"career_progression":10,

"startup_mindset":5,

"education":5,

"certifications":5

}
```

```python
JD_PROFILE={

"skills":jd_skills,

"minimum_experience":int(matches[0]) if matches else 0,

"mandatory":mandatory_sentences,

"preferred":preferred_sentences,

"negative":negative,

"production":production,

"startup":startup,

"weights":weights

}
```

```python
with open(

"jd_profile.json",

"w"

) as f:

    json.dump(

        JD_PROFILE,

        f,

        indent=4

    )

print("JD Profile Saved")
```
