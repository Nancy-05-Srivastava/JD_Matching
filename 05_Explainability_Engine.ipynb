```python
import pandas as pd
```

```python
ranking = pd.read_csv("ranked_candidates.csv")

reasoning = pd.read_csv("candidate_reasoning.csv")
```

```python
submission = ranking.merge(

    reasoning[["candidate_id","reasoning"]],

    on="candidate_id",

    how="left"

)
```

```python
submission = submission.head(100).copy()
```

```python
submission["rank"] = range(

    1,

    len(submission)+1

)
```

```python
submission = submission[[

    "candidate_id",

    "rank",

    "ranking_score",

    "reasoning"

]]
```

```python
submission.rename(

    columns={

        "ranking_score":"score"

    },

    inplace=True

)
```

```python
print("Rows :", len(submission))

print("Unique IDs :", submission["candidate_id"].nunique())

print("Ranks :", submission["rank"].min(),

      submission["rank"].max())

print("Duplicate IDs :", submission["candidate_id"].duplicated().sum())

print("Duplicate Ranks :", submission["rank"].duplicated().sum())
```

```python
print(

submission["score"].is_monotonic_decreasing

)
```

```python
submission.to_csv(

    "submission.csv",

    index=False,

    encoding="utf-8"

)

print("submission.csv generated successfully.")
```
